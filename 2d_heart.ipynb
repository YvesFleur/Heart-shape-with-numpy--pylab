```python
import numpy as np
import pylab as plb

""" This function was found here: https://www.wolframalpha.com/input?i=heart+curve+5 """

t = np.linspace(0, 2 * plb.pi, 1000) # create 1000 values of t between 0 and 2*pi

x = 16 * (plb.sin(t) ** 3)
y = 13 * plb.cos(t) - 5 * plb.cos(2*t) - 2 * plb.cos(3*t) - plb.cos(4*t)

plb.plot(x, y, color='red')
plb.show()
```
